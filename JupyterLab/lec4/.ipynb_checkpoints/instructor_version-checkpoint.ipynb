* assign random points in 2d and plot

```python

nsamp=1000
X = np.random.random(size=3*nsamp).reshape((nsamp,3))

PltErr(X[:,0],X[:,1],
#      xlim=(0,2.5), #--- range in x
#      ylim=(-2,2), #--- range in y
      attrs={'fmt':'.'}, #--- solid line
      xstr='$x$', #--- x-axis label
      ystr='$y$', #--- y-axis label
      )


```

* pairwise distances

```python
def Distance(xi,xj):
    xij = xj - xi
    rijsq = xij[0]*xij[0]+xij[1]*xij[1]+xij[2]*xij[2]
    return rijsq**0.5
    
#--- loop over pair ij and compute pairwise distance
count = 0
rij = {}
for i in range( nsamp ):
    for j in range( i+1, nsamp ):
        rij[ count ] = Distance(X[i,:],X[j,:])
        count += 1

PltErr(rij.values(),np.zeros(len(rij)),
#      xlim=(0,2.5), #--- range in x
#      ylim=(-2,2), #--- range in y
      attrs={'fmt':'x','alpha':.01}, #--- solid line
      xstr='$r_{ij}$', #--- x-axis label
      )
```

* plot $N(r)$

```python
value = np.array(list(rij.values()))
#
rmin = value.min()
rmax = value.max()
bins = np.logspace(np.log10(rmin),np.log10(rmax),32)
#
hist, bin_edges = np.histogram( value, bins = bins )
rmean = (bin_edges[:-1]*bin_edges[1:])**0.5 #--- center
#dr = bin_edges[1]-bin_edges[0]


#volume = 4.0*np.pi*rmax**3/3
#rho = (len( value ) + 1) / volume
        
#hist = hist.astype(float)    
#hist /= 4*np.pi*rmean*rmean*dr
#hist /= rho  


ax = PltErr(rmean,np.cumsum(hist),
#      xlim=(0,2.5), #--- range in x
#      ylim=(0,4), #--- range in y
      attrs={'fmt':'-o'}, #--- solid line
      xstr='$r$', #--- x-axis label
      ystr='$N(r)$', #--- y-axis label
      xscale='log',
      yscale='log',
      Plot=False,      
      )
df=3
PltErr(rmean, 10**6*rmean**df,
       ax=ax,
       attrs={'fmt':'-.r','label':'$r^%s$'%df},
       Plot=True,
       legend=True,
      )
```
